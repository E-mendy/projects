# Modeling Time series

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data1['Non_Traditional_Travellers'], lags=100)  
plt.show()


plot_pacf(data1['Non_Traditional_Travellers'])
plt.show()



# this implies non stationary, random walk, autocorrelation
```

In the ACF chart we see only three significant spikes that are on the positive, meaning there is a significant correlation up to 3 lags

# Explore the Seasonality (time-series)

```python
#plotting he seasonal decomposition for Non Traditional Travellers
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data1["Non_Traditional_Travellers"].dropna(), period = 10 )

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
plt.figure(figsize=(16,8))
plt.subplot(411) # (nrows, ncols, index). 
plt.title("Decompostion with freq=10 years")
plt.plot(data1["Non_Traditional_Travellers"].dropna(), label="original")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

We can see that the trend and seasonality information extracted from the series does seem reasonable. The residuals are also interesting, showing periods of high variability in the early and later years of the series.
we can see in the trend there has been an increase over the years till the 2020 towards 2021 which was during the pandemic perod.
In the seasonality the peak period is always the endof the year towards the begining of the year for example is October down to January as seen above till towards the pandemic period.



```python
#Plotting the seasonal Decompostion for Traditional Travellers
plt.figure(figsize=(16,8))
plt.subplot(411) # (nrows, ncols, index). 
plt.title("Decompostion with freq=10 years")
plt.plot(data1["Traditional_Travellers"].dropna(), label="original")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```
